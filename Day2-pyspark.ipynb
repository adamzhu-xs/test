```python
file = 'data/spark/Cartier+for+WinnersCurse.csv'
```

```python
sdf = spark.read.format('csv').option("header", "true").option("inferSchema", "true").load(file)
```

```python
sdf1 = sdf.filter(sdf.price > 1000).collect()
```

```python
sdf.filter("bidderrate == 5").show()
```

```python
sdf.filter(F.col('bidderrate') == 5).show()
```

```python
sdf.filter(F.col('bidder').isin('adammurry','cberdouare')).show()
```

```python
sdf.select('auctionid').distinct().count()
```

```python
sdf.select('auctionid').distinct().explain(extended=True)
```

```python
df3 = sdf.withColumn('new_price', F.col('price') * 2)
```

```python
df3.show()
```

```python
sdf.agg(F.max('bid'), F.count('bidder')).show()
```

```python
sdf.groupBy('auctionid').agg(F.sum('bid').alias('bids'), F.count('bidder').alias('cb')).select('auctionid', 'bids', 'cb').show()
```

```python
sdf.groupBy('auctionid').agg(F.sum('bid').alias('bids'), F.count('bidder').alias('cb')).show()
```

```python
sdf.summary().select('summary', 'price').show()
```

```python
sdf.stat.crosstab('bidder', 'auctionid').show()
```

```python
sdf.registerTempTable("auctions")
spark.sql("select count(*) as count from auctions").show()
```

```python

spark.sql("select distinct(auctionid) as id from auctions").show()
```

```python
from pyspark.sql.types import *
slen = F.udf(lambda s: len(s), IntegerType())
spark.udf.register('slen', slen)
spark.sql("select slen(bidder) as len from auctions").show()
```
